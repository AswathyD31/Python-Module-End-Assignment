# Analysis Tasks:

## 1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.

```python
teams = df['Team'].unique()
print(teams)
```

```python
# Calculate the distribution of employees across each team
team_distribution = df['Team'].value_counts()
team_distribution
```

```python
total_employees=len(df)
print(total_employees)
```

```python
# Calculate the percentage split relative to the total number of employees
team_distribution_percentage = (team_distribution / df.shape[0]) * 100
team_distribution_percentage
```

```python
team_distribution_df = pd.DataFrame({
    'Count': team_distribution,
    'Percentage': team_distribution_percentage
}).reset_index().rename(columns={'index': 'Team'})
```

```python
team_distribution_df
```

# Graphical Representation:

```python
#Visualizing the results.
plt.figure(figsize=(12, 8))
sns.barplot(x='Count', y='Team', hue='Team', data=team_distribution_df, palette='viridis', dodge=False, legend=False)
plt.xlabel('Number of Employees')
plt.ylabel('Team')
plt.title('Distribution of Employees Across Teams')
plt.show()
```

```python
#Total number of teams.
print("Total number of teams=",len(team_distribution))
```

# Graphical Representation:

```python
#Visualizing the results.
plt.figure(figsize=(55,25),dpi=30)
x=team_distribution_percentage.index
y=team_distribution_percentage.values
plt.bar(x, y, color=['red','purple'],width=0.5)
plt.xticks(fontsize=35)
plt.xticks(rotation=90)
plt.yticks(fontsize=35)
plt.title('Percentage split of employees in each team',size=45)
plt.xlabel('Team',size=45)
plt.ylabel('Percentage',size=45)
plt.show()
```

#### #Data Story: The distribution of employees across teams is fairly uniform, with a few teams having slightly more employees.
